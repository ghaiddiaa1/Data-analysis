```python
#Import Libraies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Laod And Read Data 

df = pd.read_csv('/Users/ghaidaa/Desktop/Titanic/Titanic.csv')
```

```python
df.head(5)
```

```python
# What is the total number of passengers in the dataset?

total_passengers = sum(df["Survived"].value_counts())
total_passengers
```

```python
# How many passengers survived the disaster?

total_survived = df["Survived"].value_counts().plot(kind="pie") 
# 0 = don't survived, 1 = survived

total_survived
```

```python
df.groupby(["Sex", "Survived"])["Survived"].value_counts()
```

```python
sns.countplot(x="Survived", hue="Sex", data=df)
```

```python
# What is the age of the youngest passenger?

df["Age"].min()
```

```python
# What is the age of the oldest passenger?

df["Age"].max()
```

```python
df["Sex"].value_counts()
```

```python
# How many passengers were in first class?
# How many passengers were in second class?
# How many passengers were in third class?

df["Pclass"].value_counts()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df.Fare, y=df.Age)
```

```python
# What is the lowest fare paid by a passenger?

df["Fare"].min()
```

```python
# We have people of all age group similarly distributed across three classes with elederly people more in class 1

plt.figure(figsize=(10,4))
sns.set_theme(style='ticks')
sns.stripplot(x=df.Pclass, y=df.Age, palette='Set1', linewidth=0.6, jitter= 0.3)
plt.show()
```

```python
# How many passengers had siblings or spouses on board?

df["SibSp"].value_counts()
```

```python
plt.figure(figsize=(16,8))
sns.stripplot(x=df.SibSp, y = df.Age ,  palette='pastel', size = 7, linewidth=1,jitter=0.2)
```
